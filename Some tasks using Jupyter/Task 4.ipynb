```python
import user_agents
```

```python
def extract_data(data):
    ua = user_agents.parse(data)
    return pd.Series((ua.browser, ua.os ,ua.device))

df[['Browser','OS','Device']] = df["UserAgent"].head(50).apply(extract_data)
```

# 2) part 2

```python
from ip2geotools.databases.noncommercial import DbIpCity
```

```python
def get_lat_long(IP):
    IPaddress = str(IP)
    from_ipaddress = DbIpCity.get(ip_address = IPaddress , api_key = 'free')
    return pd.Series((from_ipaddress.latitude, from_ipaddress.longitude))
```

```python
df[['Latitude','Longtitude']] = df['IP'].head(50).apply(get_lat_long)
```
