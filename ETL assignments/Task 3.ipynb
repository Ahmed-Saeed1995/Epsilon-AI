# 1) part 1

```python
def extract_title(x):
    x = str(x)
    if 'Mr' in x:
        return 'Mr'
    elif 'Mrs' in x:
        return 'Mrs'
    elif 'Miss' in x:
        return 'Miss'
    
df['Title'] = df['Name'].apply(extract_title)
```
